## Basic plotting using matplotlib

```python
import matplotlib.pyplot as plt

## uncomment the following to test the help function
#help(plt.plot)
```

```python
squares = [1, 3, 9, 16, 25]

plt.figure()
plt.plot(squares)
plt.ticklabel_format(axis='both', style='sci')
plt.show()
```

## Colormaps
cmap

https://matplotlib.org/users/colormaps.html

```python
x_values = list(range(0,1000))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, edgecolor='none',
            c=y_values, cmap=plt.cm.prism, s=50)

## control tick labels
plt.ticklabel_format(axis='both', style='sci', scilimits=(0, 4))
plt.show()

plt.scatter(x_values, y_values, edgecolor='none',
            c=y_values, cmap=plt.cm.prism, s=100)
plt.show()
```

## Output File Format and Resolution

Formats: png, pdf, ps, eps and svg

Recommended formats: svg and pdf (don't need to worry about setting the resolution then using DPI).

Sometimes you might need a png file. The "dots-per-inch" (dpi) is important for this. For print material, you need a minimum of 300 dpi.


```python
## colormap - https://matplotlib.org/tutorials/colors/colormaps.html
## save figure as a png-formatted file
## [wdith, height] size is in inches
plt.figure(figsize=[6.4, 4.8], dpi=300)
plt.scatter(x_values, y_values, edgecolor='none',
            c=y_values, cmap=plt.cm.plasma, s=50)
plt.savefig('squares.png', bbox_inches='tight')
plt.show()
```

## Overlay of two data sets

Also covered:
1. linesytle
2. legend
3. grid

```python
## numpy: a "fundamental package for scientific computing"
import numpy as np

## arrange - returns evenly spaced values with a given interval.
t = np.arange(0.0, 1.0, 0.01)

## create two data sets
s = 1 + np.sin(2*np.pi*t)
c = 1 + np.cos(2*np.pi*t)

## see https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot

## possible linestyles are ':' (dots), '-.' (dash-dot),
##                         '--' (dash) and '-' (line)

plt.plot(t, s, color='black', linewidth=5, linestyle='--',
         label='Human Signal')
plt.plot(t, c, color='#E82C0C', linewidth=5, linestyle=':',
         label='Alien Signal')

legend = plt.legend(loc='upper center', shadow=True, fontsize='x-large')

plt.xlabel('X-Axis (Unit)')
plt.ylabel('Y-Axis (Unit)')
plt.title('TITLE')
plt.grid(True)

#plt.savefig("simple_matplotlib.png", bbox_inches='tight', dpi=300)
#plt.savefig("simple_matplotlib.svg")
plt.show()
```

## Formatting Issues
1. tick label rotation = n degrees
2. labelpad = n pts (space between tick labels and axis label
3. tickmark width and length (via tick_params)
4. axes line width (via rc)
5. fontsize = n pts
6. fontstyle = ['normal' | 'italic' | 'oblique']
7. fontweight = ['normal' | 'bold' | 'heavy' | 'light']


```python
plt.plot(t, s, color='black', linewidth=5, linestyle='--',
         label='Human Signal')
plt.plot(t, c, color='#E82C0C', linewidth=5, linestyle=':',
         label='Alien Signal')

legend = plt.legend(loc='upper center', shadow=True, frameon=True, fontsize='x-large')

plt.xlabel('X-Axis (Unit)', fontsize=18, fontweight='bold', labelpad=15)
plt.yticks(fontsize=18)

## axis linewidth
plt.rc('axes', linewidth=5)

plt.ylabel('Y-Axis (Unit)', fontsize=18, fontweight='bold', labelpad=15)
plt.xticks(rotation='90', fontsize=18)

# tick mark's length, width and color
plt.tick_params('both', length=10, width=5, colors='#E82C0C')

plt.title('Legal vs. Illegal Space Aliens',
          fontsize=18, fontstyle='italic', fontweight='bold')

plt.show()
```

## Controlling tick mark range and minor ticks


```python
from matplotlib.ticker import AutoMinorLocator

plt.plot(t, s, color='black', linewidth=5, linestyle='--',
         label='Human Signal')
plt.plot(t, c, color='#E82C0C', linewidth=5, linestyle=':',
         label='Alien Signal')

legend = plt.legend(loc='upper center', shadow=False, frameon=True, fontsize='x-large')

plt.xlabel('X-Axis (Unit)', fontsize=18, fontweight='bold', labelpad=15)
plt.yticks(fontsize=18)

## axis linewidth
plt.rc('axes', linewidth=5)

plt.ylabel('Y-Axis (Unit)', fontsize=18, fontweight='bold', labelpad=15)
plt.xticks(rotation='90', fontsize=18)

# tick mark's length, width and color
plt.tick_params('both', length=10, width=5, colors='#E82C0C')

plt.title('Legal vs. Illegal Space Aliens',
          fontsize=18, fontstyle='italic', fontweight='bold')

plt.yticks(np.arange(min(s), max(s)+1, 1.0))

plt.minorticks_on()
plt.rc('axes', linewidth=3)
plt.tick_params(axis='x', which='minor', direction='out')
plt.tick_params(which='both', width=3)
plt.tick_params(which='major', length=7)
plt.tick_params(which='minor', length=4, color='black')

plt.show()
```

## Reading CSV-Formatted Files

data_3d.csv:
1. hearder first line
2. 28 data rows
3. 3 data columns (Time, Exp and Theory)

```python
!head data_3d.csv --lines=10 ## works for linux's bash only
```

```python
import csv

time = []
exp = []
sim = []

with open('data_3d.csv') as file:
    read_it = csv.reader(file, delimiter=',')
    next(read_it)
    for row in read_it:
        time.append((row[0]))
        exp.append(float(row[1]))
        sim.append(float(row[2]))

print('Time: {0}\n'.format(time))
print('Exp: {0}\n'.format(exp))
print('Sim: {0}\n'.format(sim))
```

### Scatterplot

```python
# Need to change plot dimensions to get x-axis label readable
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(15,5))

plt.plot(time, exp, linewidth=5, linestyle='-', label='Experimental')
plt.plot(time, sim, linewidth=5, linestyle='--', label='Simulated')

plt.xlabel('Time (seconds)')
plt.ylabel('Y-Axis (Unit)')
plt.title('Experimental and Simulated Results')
plt.grid(True)
plt.show()
```

## Bar plot

bar(x, height, width=0.8, bottom=None, \*, align='center', data=None, **kwargs)

x : sequence of scalars -- The x coordinates of the bars. See also align for the alignment of the bars to the coordinates.

height : scalar or sequence of scalars -- The height(s) of the bars.

width : scalar or array-like, optional -- The width(s) of the bars (default: 0.8).

```python
exp_average = np.mean(exp)
sim_average = np.mean(sim)

# sample standard deviation via ddof=1 (reproduces libreoffice stdev)
exp_std = np.std(exp, ddof=1)
sim_std = np.std(sim, ddof=1)

# round to two decimal places
print('Experimental Average: {0:.2f}'.format(exp_average))
print('Simulated Average: {0:.2f}'.format(sim_average))
print('Experimental Standard Deviation: {0:.2f}'.format(exp_std))
print('Simulated Standard Deviation: {0:.2f}'.format(sim_std))
```

```python
## within each category (i.e. 1 in exp; 1 in sim) -- change
##     to 2 and see what happens
n_groups = 1  

means = [exp_average, sim_average]
standard_dev = [exp_std, sim_std]

index = np.arange(n_groups)
bar_width = 0.15

## the following changes the error bar opacity
opacity = 0.4   # changes color
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, exp_average, bar_width,
                 color='C1',
                 alpha=opacity,
                 yerr=exp_std,
                 ## errorbar - error_kw
                 error_kw={'ecolor': ['C1'], 'alpha':'0.3',
                           'elinewidth':4},
                 label='Experimental')
rects2 = plt.bar(index + bar_width, sim_average, bar_width,
                 color='C6',
                 alpha=opacity,
                 yerr=sim_std,
                 error_kw={'ecolor': ['C6'], 'alpha':'0.3',
                           'elinewidth':4},
                 label='Simulated')
plt.xlabel('Methodology')
plt.ylabel('Values')
plt.title('Experimental and Simulated Results')
plt.tick_params(
    axis='x',         ## changes apply to the x-axis (x, y, both)
    which='both',     ## both major and minor ticks (major, minor, both)
    bottom=False,     ## ticks along the bottom edge are off
    top=False,        ## ticks along the top edge are off
    labelbottom=False) ## labels along the bottom edge are off
plt.legend()
plt.grid(False)
plt.show()
```

```python
## bar graph - different x-axis labels
plt.figure()
bars = ('Cis', '30.0', 'Gauche', '90.0', '120.0', '150.0', 'Anti')


position=[0, 1, 2, 3, 4, 5, 6]
## could have used a list instead of a tuple, or 
##     'pos = np.arange(len(bars))'
observable=(0.5, 1.0, 2.0, 3.0, 2.0, 1.0, 0.5)

plt.bar(position, observable, color=['black', 'black', 'green',
                                     'black', 'black'])
plt.xticks(position, bars)
plt.show()
```

## Subplots

subplot(nrows, ncols, index)

```python
# for simplicity, let's define some functions that plot
def plot1():
    plt.style.use('seaborn-whitegrid')
    plt.plot(time, exp, linewidth=5, linestyle='-', label='Experimental')
    plt.plot(time, sim, linewidth=5, linestyle='--', label='Simulated')
    plt.yticks(fontsize=14)
    ## rotate tick lables in n degrees
    plt.xticks(time, rotation='90', fontsize=14)
    plt.xlabel('Time (seconds)', fontsize=18, fontweight='bold')
    plt.ylabel('Y-Axis (Unit)', fontsize=18, fontweight='bold')
    plt.title('Experimental and Simulated Results', fontsize=18,
              fontweight='bold')
    plt.grid(False)
    plt.legend()

def plot2():
    plt.style.use('seaborn-whitegrid')
    x_values = [1, 2, 3, 4, 7]
    y_values = [1, 4, 9, 16, 49]
    plt.scatter(x_values, y_values, edgecolor='dimgray', linewidth='5',
                facecolor='C6', alpha=0.5, s=500, marker='s',
                label='Tests')
    plt.legend()
    
def plot3():
    plt.xkcd()
    rects1 = plt.bar(index, exp_average, bar_width,
                 color='C1',
                 alpha=opacity,
                 yerr=exp_std,
                 error_kw={'ecolor': ['C1'], 'alpha':'0.3',
                           'elinewidth':4},
                 label='Experimental')
    rects2 = plt.bar(index + bar_width, sim_average, bar_width,
                 color='C2',
                 alpha=opacity,
                 yerr=sim_std,
                 error_kw={'ecolor': ['C2'], 'alpha':'0.3',
                           'elinewidth':4},
                 label='Simulated')
    plt.xlabel('Methodology')
    plt.ylabel('Values')
    plt.title('Results by Methodology')
    plt.title('Experimental and Simulated Results')
    plt.tick_params(
        axis='x',
        which='both',
        bottom=False,
        top=False,
        labelbottom=False)
    plt.legend()
```

```python
## needed for the plot1
plt.figure(figsize=(15,5))

## (2r x 1c grid, first subplot position)
plt.subplot(2, 1, 1)
plot1()

## (2r x 1c grid, second subplot position)
plt.subplot(2, 1, 2)
plot2()

## addes an extra height padding between the plots
plt.tight_layout(h_pad=3.0)
plt.show()
```

## Dealing with Regional (i.e locale) numbers (e.g. DE =  1.000,55)

```python
import pandas as pd #Another way to read in csv files

import locale
## set to German locale to get comma decimal separater
## to see a list in Linux bash shell: 'locale -a'
locale.setlocale(locale.LC_ALL, ('de_DE', 'UTF-8'))
```

```python
df = pd.read_csv("data_eu.csv", header=0, sep=';', decimal='.',
                 thousands='.')
print(df)
```

```python
df['sum'] = df['Theory'] + df['Value']
print(df)
```

```python
plt.rcdefaults()
plt.style.use('seaborn-whitegrid')

plt.figure(figsize=(15,5))
plt.xticks(rotation='90')
plt.scatter('Time', 'Value', data=df, edgecolor='dimgray',
            linewidth='5', facecolor='C6',
            alpha=0.5, s=500, marker='s', label='Tests')
plt.show()
```

```python
df = pd.read_csv("data_eu.csv", sep=';',
     converters={
        'Time':   lambda x: float(x.replace('.','').replace(',','.')),
        'Exp':    lambda x: float(x.replace('.','').replace(',','.')),
        'Theory': lambda x: float(x.replace('.','').replace(',','.')),
        'Value':  lambda x: float(x.replace('.','').replace(',','.'))})
print(df)
```

```python
df['sum'] = df['Theory'] + df['Value']
print(df)
```

```python
from matplotlib.ticker import StrMethodFormatter ## for formating ticks

locale.setlocale(locale.LC_ALL, ('en_US', 'UTF-8'))
plt.style.use('seaborn-whitegrid')

plt.figure(figsize=(15,5), dpi= 300, facecolor='w', edgecolor='k')
#plt.figure(figsize=(15,5))
plt.xticks(rotation='90')
plt.scatter('Time', 'Value', data=df, edgecolor='dimgray', linewidth='5',
            facecolor='C6', alpha=0.5, s=500, marker='s', label='Tests')

plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.2f}'))
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}'))

plt.show()
```
