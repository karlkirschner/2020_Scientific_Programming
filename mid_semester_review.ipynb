## Scientific Programming

**Definition**
1. Programming whose goal is for scientific usage (e.g. workflows, data analysis) and visualization.
2. Learning to program is an academic, scholarly manner (i.e. wissenschaftliches Arbeit).

**3 Ways to Think About It**
1. **Usage**: to perform mathematics (simple numerical computations to some complex math models)
2. **Practice**: to create while maintaining good scholarship
    - knowing what is state-of-the-art
    - careful, clear and supportive
    - "A machine requires precise instructions, which users often fail to supply"[1]
3. **Target**: to support science (doing research, data support and analysis)
    - "Scientists commonly use languages such as Python and R to conduct and **automate analyses**, because in this way they can **speed data crunching**, **increase reproducibility**, protect data from accidental deletion or alteration and handle data sets that would overwhelm commercial applications."[1]
    - Create workflows to help do the research
    - Create simulations (increasingly becoming more important in research)
        - exploratory: for understanding raw data
        - supportive: for strengthening interpretations of the data
        - predictive: creating new ideas
 
[1] Baker, Monya. "Scientific computing: code alert." Nature 541, no. 7638 (2017): 563-565.

```python
'''
Scientific Programming with Python
Assignment 1, Task 2
Author: ABC (abc)
Date: November 2, 2020

Calculate the nonbounded interaction energy between two argon atoms
'''

atom_distance = 3.5
epsilon = 0.0103
sigma = 3.4

potential_energy = 4 * epsilon * ((sigma/atom_distance)**12 - (sigma/atom_distance)**6)

print('The non bounded interaction energy between two argon atoms that are 3.5A apart is {0} kJ/mol.'.format(potential_energy))
```

```python
def eV_2_J(eV: float=None) -> float:
    '''eV to J conversion '''
    
    if eV == None:
        print("An eV value was not provided.")
        return None
    else:
        return eV*1.602176565e-19


def eV_2_J_mol(eV: float=None) -> float:
    '''eV to J/mol conversion '''
    
    if eV == None:
        sys.exit("An eV value was not provided.")
    else:
        return eV*96487


def eV_2_kJ_mol(eV: float=None) -> float:
    '''eV to kJ/mol conversion '''
    
    if eV == None:
        sys.exit("An eV value was not provided.")
    else:
        return eV*96.487

    
def lj_energy(distance: float=None, epsilon: float=None, sigma: float=None) -> float:
    """
    Implementation of the 6-12 Lennard-Jones potential to calculate the energy of the interaction.

    Parameter Inputs:
    distance: Distance between two particles (Å)
    epsilon: Negative of the potential energy at the equilibrium bond length (eV)
    sigma:   Distance at which the potential energy is zero (Å)

    Returns:
    Energy of the Lennard-Jones potential (eV)
    """

    if distance == None:
        sys.exit("A distance was not provided.")
    elif epsilon == None:
        sys.exit("An epsilon value was not provided.")
    elif sigma == None:
        sys.exit("A sigma value was not provided.")
    else:
        return (4*epsilon*((sigma / distance)**12 - (sigma / distance)**6))


potential_energy = lj_energy(distance=3.5, epsilon=0.0103, sigma=3.40)

print('The potential energy of Ar---Ar, separated by '
      '3.5 Å is {0:.1e} J.'.format(eV_2_J(potential_energy)))

print('The potential energy of Ar---Ar, separated by '
      '3.5 Å is {0:.1e} J/mol.'.format(eV_2_J_mol(potential_energy)))

print('The potential energy of Ar---Ar, separated by '
      '3.5 Å is {0:.1e} kJ/mol.'.format(eV_2_kJ_mol(potential_energy)))
```

```python
import csv
import statistics

names = []
number_of_pieces = []

## read specific columns in as specified types
with open('lego_sets.csv') as file:
    read_it = csv.reader(file, delimiter=',')
    next(read_it)
    for row in read_it:
        names.append(str(row[1]))
        number_of_pieces.append(int(row[4]))

print('Name length: {0}\n'.format(len(names)))
print('Name: {0}'.format(names))
```

```python
maximum = max(number_of_pieces)
maximum
```

```python
statistics.mean(number_of_pieces)
```

```python
large_NoPs = []
long_names = []

index = 0

for size in number_of_pieces:
    if size >= 1000:
        large_NoPs.append(number_of_pieces[index])
        long_names.append(names[index])
        
        print(index, names[index], size)
    index += 1
    
name_size = zip(long_names, large_NoPs)
        
with open("lego_sets_large.csv","w+") as csvfile:
    writer = csv.writer(csvfile)
    for row in name_size:
        writer.writerow(row)
```
