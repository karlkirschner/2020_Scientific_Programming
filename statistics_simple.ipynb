## The Statistics Library
- Relaively Simple (compared to numpy or scipy)
    - low number of other library dependencies (i.e. more "usable" by the community)
- Relaively Slow (compared to numpy) - use with smaller data sets
- More precise (compared to numpy) - when high precision is important

## Mean, median, standard deviation and variance
- Measures of the center in a numerical data
    - Mean (i.e. average): can be influenced by several outliers
    - Median: the middle value (less influenced by outliers)
- Standard Deviation:  a measure of the data's uncertainty
- Variance: a measure of the data's spread

## Calculate some common statistics (and show on plot)

### Data Set 1

```python
print('Mean:', statistics.mean(data_1))
print('Mediam', statistics.median(data_1))
print('StDev:', statistics.stdev(data_1))
print('Variance', statistics.variance(data_1))
```

```python
plt.figure()
plt.xlim(0, 9)
plt.ylim(30, 120)

plt.axhline(y=67, color='red')
plt.axhline(y=66, color='green')

plt.plot(data_1, marker='.', markersize=10)
plt.show()
```

## How might the above code been done better?

- reduce possible inclusion of errors
- include the practice of significant figures

```python
def do_statistics(working_data=None):
    print('Mean:', round(statistics.mean(working_data)))
    print('Mediam', round(statistics.median(working_data)))
    print('StDev:', round(statistics.stdev(working_data)))
    print('Variance', round(statistics.variance(working_data)))

do_statistics(data_1)
```

```python
## Same idea, but with different print formatting
## Could even be better, by reducing the redundant information! (being more scientific)
def do_statistics(working_data=None):
    print("{0}'s Mean: {1}".format(working_data, round(statistics.mean(working_data))))
    print("{0}'s Median {1}".format(working_data, round(statistics.median(working_data))))
    print("{0}'s StDev: {1}".format(working_data, round(statistics.stdev(working_data))))
    print("{0}'s Variance {1}".format(working_data, round(statistics.variance(working_data))))
    
do_statistics(data_2)
```
